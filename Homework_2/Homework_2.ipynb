```python
import numpy as np 
import pandas as pd
from numpy.linalg import inv
from numpy import dot
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt 
%pylab inline

from mpl_toolkits.mplot3d import Axes3D  

h31 = pd.read_csv('data/H31.csv', index_col = 0)

# x3 is the dummy variable
```

```python
y = h31.y 
x = h31[['x1', 'x2', 'x3']]
w = dot(inv(dot(x.T, x)), dot(x.T, y))
w
```

```python
reg = smf.ols('y ~ x1 + x2 + x3', h31).fit()
reg.params
```

```python
fig = plt.figure(figsize = (12,10))  # Create a plot with customized figure size (width, height)
ax = fig.gca(projection='3d')  # Extract the axis for further operations
ax.scatter(h31.x1, h31.x2, h31.y, s = 100,  c = 'r')
# Scatter plot, "c" stands for color, and "s" determines the marker size

# Generate coordinates for hyperplane
radius = 5
x1, x2 = np.meshgrid(range(-radius,radius), range(-radius,radius))
fit_y = reg.params[0] + reg.params[1] * x1 + reg.params[2] * x2

# alpha (between 0 and 1) controls the transparency with 0 being totally transparent
ax.plot_surface(x1, x2, fit_y, color='c', alpha = 0.4)
# Set labels and fontsize
ax.set_xlabel('$x_1$', fontsize = 20)
ax.set_ylabel('$x_2$', fontsize = 20)
ax.set_zlabel('$y$', fontsize = 20)

ax.view_init(azim = 60)   # Controls the viewing angle
```

```python
h33 = pd.read_csv('data/H33.csv')
```

```python
#a) 
plt.plot(h33.x1, h33.y, '.');
```

```python
#b)
h33['x2'] = h33.x1**2
h33['x3'] = h33.x1**3
h33['x4'] = h33.x1**4
```

```python
#c)
fit1 = smf.ols('y ~ x1', h33).fit()
fit2 = smf.ols('y ~ x1 + x2', h33).fit()
fit3 = smf.ols('y ~ x1 + x2 + x3', h33).fit()
fit4 = smf.ols('y ~ x1 + x2 + x3 + x4', h33).fit()
```

```python
rsq = [fit1.rsquared, fit2.rsquared, fit3.rsquared, fit4.rsquared]
coef = [fit1.params, fit2.params, fit3.params, fit4.params]
```

```python
#d) 
print('rsquared values for fitted polynomials of degree 1,2,3,4 are', rsq)
plt.plot([1,2,3,4], rsq, '-');
plt.xlabel('degrees of polynomial')
plt.ylabel('r squared value');
```

```python
def plot_fit(location, fit, degree):
    fig.add_subplot(location)
    plt.plot(h33.x1, h33.y, '.')
    plt.plot(h33.x1, fit.predict(), label='fitted curve')
    plt.plot(h33.x1, 9*h33.x1**2 - h33.x1 + 1, label='True curve')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('M = '+ str(degree))
    plt.legend()
```

```python
fig = plt.figure(figsize = (20,5))
plot_fit(141, fit1, 1)
plot_fit(142, fit2, 2)
plot_fit(143, fit3, 3)
plot_fit(144, fit4, 4)
```

based on the plots, we can see that M = 2 and 3 are the better fits
